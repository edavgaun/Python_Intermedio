# Repaso regresi√≥n lineal

```python
from sklearn.linear_model import LinearRegression

df_males = data[data['Gender'] == 'Male']
lr_males = LinearRegression()

df_females = data[data['Gender'] == 'Female']
lr_females = LinearRegression()
```

```python
lr_males.fit(df_males[['Altura']], df_males['Peso'])
lr_females.fit(df_females[['Altura']], df_females['Peso'])
```

```python
print("-"*20)
print("Hombres")
print("Ordenada\t",lr_males.intercept_)
print("Pendiente\t",lr_males.coef_)
print("-"*20)
print("Pearson\t", df_males.corr())
print("-"*20)
print("Mujeres")
print("Ordenada\t",lr_females.intercept_)
print("Pendiente\t",lr_females.coef_)
print("-"*20)
print("Pearson\t", df_females.corr())
print("-"*20)
```

### Hombres

$Peso=1.065(Altura)-101.831$

$R^2=0.863$

### Hombres

$Peso=1.07(Altura)-111.59$

$R^2=0.85$

```python
peso_males=lr_males.predict(df_males.Altura[:,np.newaxis])
peso_females=lr_females.predict(df_females.Altura[:,np.newaxis])
```

```python
MSE_m=sum((peso_males-df_males.Peso)**2)/5000
MSE_f=sum((peso_females-df_females.Peso)**2)/5000
```

```python
RMSE_m=np.sqrt(MSE_m)
RMSE_f=np.sqrt(MSE_f)
print(RMSE_m,RMSE_f)
```

# Regresi√≥n M√∫ltiple

```python
data.head()
```

```python
dummies=pd.get_dummies(data.Gender)
dummies.head()
```

```python
data_nueva=pd.concat([data,dummies],axis=1)
data_nueva
```

```python
data_nueva.drop(["Gender","Female"],axis=1, inplace=True)
```

```python
data_nueva.rename({"Male":"Gender"}, axis=1, inplace=True)
```

```python
data_nueva.head()
```

```python
lrm=LinearRegression()
lrm.fit(data_nueva[["Gender","Altura"]],data_nueva.Peso)
```

```python
lrm.intercept_ , lrm.coef_
```

```python
from sklearn.metrics import r2_score, mean_squared_error

r2=r2_score(data_nueva.Peso,lrm.predict(data_nueva[["Gender","Altura"]]))
MSE=mean_squared_error(data_nueva.Peso,lrm.predict(data_nueva[["Gender","Altura"]]))
```

```python
r2 , MSE , np.sqrt(MSE)
```
