```python
from sklearn.datasets import make_classification
# define dataset
X, y = make_classification(n_samples=1000, n_features=7, n_informative=4, n_redundant=3, random_state=1)
# summarize the dataset
print(X.shape, y.shape)
```

```python
import pandas as pd

df=pd.DataFrame(X)
df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

sns.pairplot(df)
plt.show()
```

```python
from sklearn.decomposition import PCA
pca=PCA(7)
pca.fit(X)
```

```python
pca.explained_variance_ratio_
```

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

sns.lineplot(
    x=np.arange(1,8), y=pca.explained_variance_ratio_.cumsum(), 
    markers=True, 
)
n=3
plt.vlines(n,0,1, colors="Red", linestyles="--")
plt.xticks(range(1,8,2))
plt.xlabel("Número de componentes principales")
plt.ylabel("Proporción acumulada de explicación de la variación")
plt.title("PCA análisis")
plt.show()
```

```python
pca=PCA(3)
pca.fit(X)
X_PCA=pca.transform(X)
X_PCA
```

```python
sns.scatterplot(x=X_PCA[:,0], y=X_PCA[:,1], hue=y)
plt.show()
```

```python
import plotly.express as px
fig = px.scatter_3d(x=X_PCA[:,0], y=X_PCA[:,1], z=X_PCA[:,2],
              color=y)
fig.show()
```
